```python
#load modules needed to carry out analysis

import pandas as pd
import datetime 
from datetime import date, timedelta
import yfinance as yf
import plotly.graph_objects as go
from autots import AutoTS

today = date.today()
d1 = today.strftime('%Y-%m-%d')
end_date = d1
d2 = date.today() - timedelta(days=720)
start_date = d2
```

```python
data = yf.download('SOL-USD', start=start_date, end=end_date, progress=False)
data.head()
```

```python
data['Date'] = data.index
data = data[["Date", "Open", "High", "Low", "Close", "Adj Close", "Volume"]]
data.reset_index(drop=True, inplace=True)
#data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')
data.head()
```

```python
#data['Date'] = pd.to_datetime(data['Date'], format='%Y%m%d')
data.info()
```

```python
data.head()
date_index = data.Date.copy()
data['date_index'] = date_index
data.dtypes
```

### Visual trend for price of Bitcoin in the last two years

```python
figure = go.Figure(data=[go.Candlestick(x=data["Date"], open=data.Open["BTC-USD"], high=data.High["BTC-USD"], low=data.Low["BTC-USD"], close=data.Close["BTC-USD"])])
figure.update_layout(title = "Bitcoin Price Analysis", xaxis_rangeslider_visible=False)
figure.show()
```

### Visual trend for price of Ethereum in the last two years

```python
fig = go.Figure(data=[go.Ohlc(
    x=data['Date'],
    open=data.Open["ETH-USD"], high=data.High["ETH-USD"],
    low=data.Low["ETH-USD"], close=data.Close["ETH-USD"],
    increasing_line_color= 'cyan', decreasing_line_color= 'gray'
)])
figure.update_layout(title="Ethereum Price Analysis")
fig.show()
```

### Visual trend for price of Solana in the last two years

```python
fig = go.Figure(data=[go.Ohlc(
    x=data['Date'],
    open=data["Open"], high=data["High"],
    low=data["Low"], close=data["Close"],
    increasing_line_color= 'green', decreasing_line_color= 'orange'
)])
fig.show()
```

#### create a model instance using the AutoTS object and create prediction for the next 30 days

```python
model = AutoTS(forecast_length=30, frequency='infer', ensemble="simple")
model = model.fit(data, date_col="Date", value_col="Close", id_col=None)
prediction = model.predict()
forecast = prediction.forecast
print(forecast)
```

```python
forecast.to_csv('solana_price')
```
